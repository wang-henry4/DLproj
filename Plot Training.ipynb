```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
root_path = "/Users/yichuan/Desktop/289/Project/DLproj/history/"
```

```python
history_dict=[]
```

```python
# read in history files
for t_layer in ["conv2_block3_out","conv3_block3_out", "conv4_block3_out", "conv5_block3_out"]:
    with open(root_path+f"{t_layer}_history", 'rb') as file_pi:
        history_dict.append(pickle.load( file_pi))
```

```python
history_dict[0]
```

```python

ax = plt.subplot(111)
t1 = np.arange(0.0, 10.0, 1)
for n in [0,1, 2, 3]:
    
    plt.plot(t1, history_dict[n]['accuracy'], label="conv%d_block3_out"%(n+2,))

leg = plt.legend(loc='best', ncol=2, mode="expand", shadow=True, fancybox=True)
leg.get_frame().set_alpha(0.5)
plt.title('Accuracy of Models')
plt.xlabel('Epoch Number')
plt.ylabel('Accuracy')

plt.show()
```

```python
ax = plt.subplot(111)
t1 = np.arange(0.0, 10.0, 1)
for n in [0,1, 2, 3]:
    
    plt.plot(t1, history_dict[n]['loss'], label="conv%d_block3_out"%(n+2,))

leg = plt.legend(loc='best', ncol=2, mode="expand", shadow=True, fancybox=True)
leg.get_frame().set_alpha(0.5)
plt.title('Loss of Models')
plt.xlabel('Epoch Number')
plt.ylabel('Loss')

plt.show()
```
